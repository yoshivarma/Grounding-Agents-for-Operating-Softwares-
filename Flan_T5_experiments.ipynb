```python
from transformers import AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
inputs = tokenizer("Please answer to the following question: Go to Setup > Payroll > Setup Tasks > Retirement Plans > Click the number next to retirement plan you want to edit > Make changes > Click Save => click[Setup], click[Payroll], click[Setup Tasks], click[Retirement Plans], search[retirement plan to edit], click[retirement plan number], input[changes], click[Save]. Go to Reports > New Search > Click Paycheck from Search list =>", return_tensors="pt").input_ids
outputs = model.generate(inputs)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
#click[Reports], click[New Search], click[Paycheck from search]
```

```python
input_ids = tokenizer("translate English to German: The house is wonderful.", return_tensors="pt").input_ids
labels = tokenizer("Das Haus ist wunderbar.", return_tensors="pt").input_ids

# the forward function automatically creates the correct decoder_input_ids
loss = model(input_ids=input_ids, labels=labels).loss
loss.item()
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
inputs = tokenizer("Please answer to the following question: Go to Transactions > Financial > Approve Journal Entries => click[Transactions], click[Financial], click[Approve Journal Entries]."
        " Go to Setup > Accounting > Setup Tasks > Opening Balances Administrator. =>", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

```python
input_ids = tokenizer("translate English to German: The house is wonderful.", return_tensors="pt").input_ids
labels = tokenizer("Das Haus ist wunderbar.", return_tensors="pt").input_ids

# the forward function automatically creates the correct decoder_input_ids
loss = model(input_ids=input_ids, labels=labels).loss
loss.item()
```
